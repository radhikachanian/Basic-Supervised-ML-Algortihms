```python
def percent_missing(df):
    percent_nan = 100* df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    
    return percent_nan
```

```python
result_nan = percent_missing(df)
result_nan
```

```python
df = pd.read_csv("AMES_Final_DF.csv")
df.head()
```

```python
df.info()
```

```python
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y ,
                                                   test_size=0.1, 
                                                    random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled_X_train = sc.fit_transform(X_train)
scaled_X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet()
```

```python
grid_values = {"alpha": [0.1,1,10, 200, 500],
               "l1_ratio" : [0,0.1,0.2,0.5,0.6,0.8,1]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

random_search_model = RandomizedSearchCV(model, 
                                   param_distributions=grid_values,
                          scoring= "neg_root_mean_squared_error"
                                   ,                                 cv=10,
                                  verbose=2,
                                        refit=True)
```

```python
random_search_model.fit(scaled_X_train,y_train)
```

```python
random_search_model.best_params_
```

```python
y_preds = random_search_model.predict(scaled_X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_preds)
rmse = np.round(mean_squared_error(y_test, y_preds))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet(l1_ratio= 1, alpha= 2)

model.fit(scaled_X_train,y_train)
y_preds = model.predict(scaled_X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_preds)
rmse = np.round(mean_squared_error(y_test, y_preds))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```
